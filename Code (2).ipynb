```python
### data splitting
```

```python
from sklearn.model_selection import train_test_split
```

```python
# seperating the features from class labels
X= df.drop('target',axis=1)
y = df['target']
```

```python
# split dataset for train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
### Feature Selection
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import ExtraTreesClassifier
import numpy as np

# Initialize two different feature selection methods
skb_selector = SelectKBest(chi2, k=5)  # SelectKBest with chi-squared test
et_selector = ExtraTreesClassifier(n_estimators=100, random_state=42)

# Fit both selectors on the training data
skb_selector.fit(X_train, y_train)
et_selector.fit(X_train, y_train)

# Get selected features from each method
selected_features_skb = X.columns[skb_selector.get_support()]
selected_features_et = X.columns[et_selector.feature_importances_.argsort()[::-1][:5]]

# Combine selected features from both methods
selected_features_hybrid = np.union1d(selected_features_skb, selected_features_et)

# Print the selected features
print(selected_features_hybrid)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense
```

```python
### Initializing the MLp model
model = Sequential()
model.add(Dense(128, input_dim=6, activation='relu'))  
model.add(Dense(64, activation='relu'))  
model.add(Dense(32, activation='relu'))  
model.add(Dense(16, activation='relu'))  
model.add(Dense(1, activation='sigmoid'))  
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#selected features
X_train_new = X_train[selected_features_hybrid]
X_test_new = X_test[selected_features_hybrid]
```

```python
# Train the model on selected features
model.fit(X_train_new, y_train, epochs=50)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
y_pred = model.predict(X_test_new)
#thresholding
y_pred = (y_pred >= 0.5).astype(int)
print("F1 score = ", f1_score(y_test,y_pred))
print("Accuracy score = ", accuracy_score(y_test,y_pred))
```
